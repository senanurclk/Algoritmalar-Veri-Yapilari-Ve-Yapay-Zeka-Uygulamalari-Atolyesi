```python
sayi = 1

while sayi<=100:
  karekok = int(sayi ** 0.5) # int yazdığı için kök 2 kök 3 ü alamayız ama kök 4 zaten 2 ye eşit olduğu için tam sayı oluyor ve onu alabiliyoruz
  if karekok ** 2 == sayi:
    print(sayi)
  sayi += 1 # sayi = sayi + 1
```

```python
sayac = 4
degisken = 0

while sayac * sayac <= 100:
  tam_kare = sayac * sayac
  if 10 <= tam_kare <= 100:
    print(tam_kare)
    degisken += 1
  sayac += 1

print("Sayı Adedi: ", degisken)
```

```python
degisken = 0

for sayac in range(1, 11): # for da sayaç artışı range ile oluyor. Şu an burada birer birer artıyor
  tam_kare = sayac * sayac
  if 10 <= tam_kare <= 100:
    print(tam_kare)
    degisken += 1

print("Sayı Adedi: ", degisken)
```
